Import Modules

```python
import llm as llModel
import media as md
import json
```

Create Images

```python
md.extractFramesFromVideo("video.mp4", everyFrames = 25)
```

```python
md.displayFrames()
images = md.getFrames()
```

```python
images
```

```python
answers = []
for i in range(len(images)):
    answer = llModel.prompt(prompt = "can you give me information about how to interact with the people inside this image?", image = images[i], returnJson=True)
    answers.append(json.loads(answer))
json_formatted_str = json.dumps(answers, indent=2)
print(json_formatted_str)
```
